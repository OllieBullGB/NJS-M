```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
simulation_params = {
    'test-runs': 5,
    'num-hosts': 20,
    'num-rsu': 1,
    'num-jammer': 1
}
```

```python
def load_and_prepare(path, solution, jammer_type):
    df = pd.read_csv(path)
    df['accuracy'] = ((df['tp'] + df['tn']) / (df['tp'] + df['fp'] + df['tn'] + df['fn'])) * 100
    df['jammerType'] = jammer_type
    df['Solution'] = solution
    return df

# Reactive jammer data
df_njs   = load_and_prepare('./dynamic/legacyResults.txt', 'Old Solution', 'reactive')
df_njsm  = load_and_prepare('./dynamic/results.txt', 'New Solution', 'reactive')
df_snr   = load_and_prepare('./dynamic/snrResults.txt', 'SNR Solution', 'reactive')
df_ml    = load_and_prepare('./dynamic/snrMLResults.txt', 'ML SNR Solution', 'reactive')

# Proactive jammer data
df_proactive_njs  = load_and_prepare('./dynamic/proactiveLegacyResults.txt', 'Old Solution', 'proactive')
df_proactive_njsm = load_and_prepare('./dynamic/proactiveResults.txt', 'New Solution', 'proactive')
df_proactive_snr  = load_and_prepare('./dynamic/proactiveSnrResults.txt', 'SNR Solution', 'proactive')
df_proactive_ml   = load_and_prepare('./dynamic/proactiveSnrMLResults.txt', 'ML SNR Solution', 'proactive')

# Combine all dataframes
df_combined = pd.concat([
    df_njs, df_njsm, df_snr, df_ml,
    df_proactive_njs, df_proactive_njsm, df_proactive_snr, df_proactive_ml
])
df_combined = df_combined[df_combined['JLE'] != -1]


num_runs = df_combined['runName'].nunique()
print(num_runs)
df_combined.head()

# Solution mapping
solution_mapping = {
    'Old Solution': 'NJS',
    'New Solution': 'NJS-M',
    'SNR Solution': 'T-SNR',
    'ML SNR Solution': 'ML SNR'
}
df_combined['Solution'] = df_combined['Solution'].map(solution_mapping)

print(num_runs)
df_combined.head()
```

```python
# Group by runName, time, Solution, and jammerType, and compute the mean JLE
df_grouped = df_combined.groupby(['runName', 'time', 'Solution', 'jammerType']).mean().reset_index()

# Pivot the dataframe for easier plotting
df_pivot = df_grouped.pivot_table(values='JLE', index='time', columns=['Solution', 'jammerType'], aggfunc='mean')

# Plot the data
plt.figure(figsize=(6, 6))

# Plot for static jammer
plt.subplot(2, 1, 1)
plt.plot(df_pivot.index, df_pivot[('NJS', 'reactive')], label='NJS', marker='o')
plt.plot(df_pivot.index, df_pivot[('NJS-M', 'reactive')], label='NJS-M', marker='o')
plt.plot(df_pivot.index, df_pivot[('T-SNR', 'reactive')], label='T-SNR', marker='o')
plt.plot(df_pivot.index, df_pivot[('ML SNR', 'reactive')], label='ML SNR', marker='o')
plt.xlabel('Time (s)')
plt.ylabel(f'Average JLE across {num_runs} runs (m)')
plt.title('Comparison of JLE Over Time (Reactive Jammer)')
plt.legend()
plt.grid(True)

# Plot for proactive jammer if it exists
if ('NJS', 'proactive') in df_pivot.columns and ('NJS-M', 'proactive') in df_pivot.columns and ('T-SNR', 'proactive') in df_pivot.columns:
    plt.subplot(2, 1, 2)
    plt.plot(df_pivot.index, df_pivot[('NJS', 'proactive')], label='NJS', marker='o')
    plt.plot(df_pivot.index, df_pivot[('NJS-M', 'proactive')], label='NJS-M', marker='o')
    plt.plot(df_pivot.index, df_pivot[('T-SNR', 'proactive')], label='T-SNR', marker='o')
    plt.plot(df_pivot.index, df_pivot[('ML SNR', 'proactive')], label='ML SNR', marker='o')
    plt.xlabel('Time (s)')
    plt.ylabel(f'Average JLE (m)')
    plt.title('Comparison of JLE Over Time (Proactive Jammer)')
    plt.legend()
    plt.grid(True)
else:
    print("No proactive runs")

plt.tight_layout()
plt.show()
```

```python
# Plot the bar chart using df_pivot
plt.figure(figsize=(6, 3))
bar_width = 0.4
solutions = list(df_pivot.columns.levels[0])
jammer_types = list(df_pivot.columns.levels[1])
index = range(len(solutions))

# Plot bars for each jammer type
for i, jammerType in enumerate(jammer_types):
    # Get mean JLE for each solution for this jammer type
    means = [df_pivot[(solution, jammerType)].mean() if (solution, jammerType) in df_pivot.columns else 0 for solution in solutions]
    plt.bar([x + i * bar_width for x in index], means, bar_width, label=jammerType)
    
    # Print the average JLE for each jammer type and solution
    print(f"Average JLE for Jammer Type '{jammerType}':")
    for solution, mean_val in zip(solutions, means):
        print(f"  {solution}: {mean_val}")

# Add labels and title
plt.xlabel('Solution')
plt.ylabel('Average JLE (m)')
plt.xticks([x + bar_width / 2 for x in index], solutions)
plt.grid(True, axis='y')
plt.legend(title='Jammer Type')

# Show the plot
plt.show()
```

```python
# Group by runName, time, Solution, and jammerType, and compute the mean accuracy
# (This is already done above, but kept for clarity)
df_grouped = df_combined.groupby(['runName', 'time', 'Solution', 'jammerType']).mean().reset_index()

# Pivot the dataframe for easier plotting
accuracy_pivot = df_grouped.pivot_table(values='accuracy', index='time', columns=['Solution', 'jammerType'], aggfunc='mean')

plt.figure(figsize=(8, 6))

# --- Static Jammer Accuracy ---
plt.subplot(2, 1, 1)
for solution in ['NJS', 'NJS-M', 'T-SNR', 'ML SNR']:
    col = (solution, 'reactive')
    if col in accuracy_pivot.columns:
        plt.plot(accuracy_pivot.index, accuracy_pivot[col], label=solution, marker='o')
plt.xlabel('Time (s)')
plt.ylabel('Average Accuracy (%)')
plt.title('Accuracy Over Time (Reactive Jammer)')
plt.ylim(0, 100)
plt.legend()
plt.grid(True)

# --- Proactive Jammer Accuracy ---
plt.subplot(2, 1, 2)
proactive_found = False
for solution in ['NJS', 'NJS-M', 'T-SNR', 'ML SNR']:
    col = (solution, 'proactive')
    if col in accuracy_pivot.columns:
        plt.plot(accuracy_pivot.index, accuracy_pivot[col], label=solution, marker='o')
        proactive_found = True
plt.xlabel('Time (s)')
plt.ylabel('Average Accuracy (%)')
plt.title('Accuracy Over Time (Proactive Jammer)')
plt.ylim(0, 100)
if proactive_found:
    plt.legend()
    plt.grid(True)
else:
    plt.text(0.5, 0.5, 'No proactive runs', ha='center', va='center', transform=plt.gca().transAxes)

plt.tight_layout()
plt.show()
```

```python
# Group by runName, time, Solution, and jammerType, and compute the mean accuracy
df_grouped = df_combined.groupby(['runName', 'time', 'Solution', 'jammerType']).mean().reset_index()

# Pivot the dataframe for easier plotting
df_pivot = df_grouped.pivot_table(values='accuracy', index='time', columns=['Solution', 'jammerType'], aggfunc='mean')

# Plot the bar chart using df_pivot
plt.figure(figsize=(6, 3))
bar_width = 0.35
index = range(len(df_pivot.columns.levels[0]))

# Plot bars for each jammer type
for i, jammerType in enumerate(df_pivot.columns.levels[1]):
    subset = df_pivot.xs(jammerType, level=1, axis=1)
    plt.bar([x + i * bar_width for x in index], subset.mean(), bar_width, label=jammerType)
    
    avg_ACC = subset.mean()
    
    # Print the average JLE for each jammer type and solution
    print(f"Average ACC for Jammer Type '{jammerType}':")
    print(avg_ACC)

# Add labels and title
plt.xlabel('Solution')
plt.ylabel('Average Accuracy (%)')
plt.xticks([x + bar_width for x in index], df_pivot.columns.levels[0])
plt.grid(True, axis='y')
plt.legend(title='Jammer Type')

# Show the plot
plt.show()
```
